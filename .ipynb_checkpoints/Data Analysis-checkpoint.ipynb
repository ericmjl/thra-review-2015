```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('responses.csv', index_col='Email address')
df
```

# Socials

```python
# Ratings for each of the socials
socials = df.iloc[:, 5:13]
socials.columns = ['sushi', 'coffee hrs', 'cupcake', 'brunch', 'thanksgiving', 'christmas tree', 'finals breakfasts', 'christmas meal']
socials.mean().plot(kind='bar')
ax = plt.gca()
ax.set_ylim(0, 5)
ax.set_ylabel('rating')
plt.setp(ax.xaxis.get_ticklabels(), rotation=30, ha='right')
plt.subplots_adjust(bottom=0.25, top=0.95, left=0.1, right=0.95)
plt.savefig('socials ratings.pdf')
```

```python
# Expectations from socials. Extract single words.
import string
exclude = set(string.punctuation)
other_exclude = set()
other_exclude.add('\n')
other_exclude.add('and')
other_exclude.add('to')
other_exclude.add('nd')
other_exclude.add('to')
other_exclude.add('I')
other_exclude.add('a')
other_exclude.add('')
other_exclude.add('i')
other_exclude.add('or')
other_exclude.add('from')

from collections import Counter
wordcount = Counter()
soc_expectations = df.iloc[:,14]
for response in soc_expectations.values:
    if type(response) == str:
        for word in response.split(' '):
            word = s = ''.join(ch.lower() for ch in word if ch not in exclude)
            if word not in other_exclude:
                wordcount[word] += 1
wordcount.most_common(10)
```

```python
# Interest Groups
intgrp = df.iloc[:,18]

grpcounter = Counter()
for response in intgrp.values:
    if type(response) == str:
        for r in response.split(', '):
            grpcounter[r] += 1
            
for k, v in grpcounter.most_common(25):
    print(k + ': ' + str(v) + '\n')
```
